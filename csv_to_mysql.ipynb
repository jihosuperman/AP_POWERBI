```python
import pandas as pd
import pymysql
import numpy as np
```

```python
# Import dataset
df = pd.read_csv('./data/dataset.csv')

# Convert np.nan to None
df['clear_date'] = pd.to_datetime(df['clear_date'])
df['posting_date'] = pd.to_datetime(df['posting_date'])
df['document_create_date'] = df['document_create_date'].apply(lambda x: pd.to_datetime(str(x), format="%Y%m%d"))
df['document_create_date.1'] = df['document_create_date.1'].apply(lambda x: pd.to_datetime(str(x), format="%Y%m%d"))
df['due_in_date'] = df['due_in_date'].apply(lambda x: pd.to_datetime(str(x), format="%Y%m%d"))
df['baseline_create_date'] = df['baseline_create_date'].apply(lambda x: pd.to_datetime(str(x), format="%Y%m%d"))

df = df.replace({np.nan: None})
df = df.replace({pd.NaT: None})

df.info()
```

```python
# Connect to MySQL DB
conn = pymysql.connect(host='localhost', user='root', password=str(input()), db ='accounting') 
curs = conn.cursor(pymysql.cursors.DictCursor)
```

```python
for i in df.columns:
    print(i, df[i].astype(str).str.len().max(), df[i].dtype)
```

```python
# Create Table
sql = '''CREATE TABLE ar( 
            business_code varchar(6), 
            cust_number varchar(12), 
            name_customer varchar(30), 
            clear_date datetime, 
            buisness_year int, 
            doc_id bigint(20), 
            posting_date datetime, 
            document_create_date datetime,
            due_in_date datetime,
            invoice_currency varchar(6),
            document_type varchar(4),
            posting_id int,
            area_business varchar(6),
            total_open_amount varchar(12),
            baseline_create_date datetime,
            cust_payment_terms varchar(8),
            invoice_id varchar(20),
            isOpen int
            ) 
      ''' 
curs.execute(sql) 
```

```python
# Insert Data to DB
for index, row in df.iterrows():
    tu = (row.business_code, row.cust_number, row.name_customer, row.clear_date, row.buisness_year, row.doc_id, row.posting_date, row.document_create_date,
          row.due_in_date, row.invoice_currency, row['document type'], row.posting_id, row.area_business, row.total_open_amount,
          row.baseline_create_date, row.cust_payment_terms, row.invoice_id, row.isOpen) 
    curs.execute("""INSERT INTO ar 
                    (business_code, cust_number, name_customer, clear_date, buisness_year, doc_id, posting_date, document_create_date,
                     due_in_date, invoice_currency, document_type, posting_id, area_business, total_open_amount,
                     baseline_create_date, cust_payment_terms, invoice_id, isOpen) 
                    VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)""",
                 tu)
    print('Complete Insert!', index)
print('Complete Insert All data!!')
conn.commit()
```
